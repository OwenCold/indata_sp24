# BoxPlot

I wanted to learn how to use a boxplot in order to showcase the upper and lower quartiles of queerphobia between sexes. I could not figure out how to change the color variables of each one.

```python
fig1, ax1 = plt.subplots()
ax1.boxplot(x = [cces.loc[cces['sex']==1]['QueerPhobia'],cces.loc[cces['sex']==2]['QueerPhobia']],
           notch = True,
           patch_artist = True)
ax1.set_title('Gender vs QueerPhobia')
plt.show()
```

```python
fig2, ax2 = plt.subplots(1,2)
ax2[0].set_title('Age Histogram')
ax2[1].set_title('Political Identity')
plt.show()
```

```python
cces_freq = cces.groupby('Evaluation').describe()
```

```python
fig3, ax3 = plt.subplots()
ax3.bar(x = ['0','1','2','3'], height = cces_freq['pew_religimp']['count'])
ax3.set_title('Evaluation Count')
plt.show()
```
