```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="darkgrid")
```

```python
sns.relplot(data=dbt, x='Age', y='Insulin')
plt.show()
```

```python
sns.relplot(data=dbt, x='Age', y='Insulin', hue='Pregnancies', size='Pregnancies', alpha=.6)
plt.show()
```

# Group 2

```python
sald = pd.read_csv("Salary Data.csv")
```

```python
sald.head()
```

```python
sns.set(style='whitegrid')
```

```python
sns.scatterplot(data=sald, x='Age', y='Salary')
plt.title('Default Scatterplot: Age vs Salary')
plt.show()
```

```python
plt.figure()
sns.scatterplot(data=sald,\
                x='Age',\
                y='Salary',\
                hue='Education Level',\
                style='Gender')
plt.title('Scatterplot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}

plt.figure()
sns.scatterplot(data=sald, x='Age', y='Salary', hue='Education Level', style='Gender', palette = education_colors)
plt.title('Scatterplot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers = {"Female": "o", "Male": "s"}

plt.figure()
sns.scatterplot(data=sald, x='Age', y='Salary', hue='Education Level', style='Gender', palette = education_colors, markers = gender_markers)
plt.title('Scatterplot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers = {"Female": "o", "Male": "s"}

plt.figure()
sns.set_style("ticks")
sns.scatterplot(data=sald, x='Age', y='Salary', hue='Education Level', style='Gender', palette = education_colors, markers = gender_markers)
plt.title('Scatterplot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers = {"Female": "o", "Male": "s"}

plt.figure(figsize=(12,8))
sns.set_style("ticks")
sns.scatterplot(data=sald, x='Age', y='Salary', hue='Education Level', style='Gender', palette = education_colors, markers = gender_markers)
plt.legend(ncol=2,frameon=False,loc="lower right")
plt.title('Scatterplot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
gender_colors = {"Female": "red", "Male": "blue"}

plt.figure(figsize=(10,8))
sns.set_style("ticks")
Bachelor=sns.scatterplot(data=sald[sald["Education Level"] == "Bachelor's"], x='Age', y='Salary', hue='Gender', palette = gender_colors)
Bachelor.set_xlim(25,50)
Bachelor.set_ylim(50000,250000)

plt.legend(frameon=False,loc="upper left")
plt.title("Scatterplot: Age vs Salary by Gender among Bachelor's Degree Holders")
plt.show()
```

```python
gender_colors = {"Female": "red", "Male": "blue"}

plt.figure(figsize=(10,8))
sns.set_style("ticks")
Master=sns.scatterplot(data=sald[sald["Education Level"] == "Master's"], x='Age', y='Salary', hue='Gender', palette = gender_colors)
Master.set_xlim(25,50)
Master.set_ylim(50000,250000)

plt.legend(frameon=False,loc="upper left")
plt.title("Scatterplot: Age vs Salary by Gender among Master's Degree Holders")
plt.show()
```

```python
gender_colors = {"Female": "red", "Male": "blue"} #mapping dictionary

plt.figure(figsize=(10,8))
sns.set_style("ticks")
phd=sns.scatterplot(data=sald[sald["Education Level"] == "PhD"], x='Age', y='Salary', hue='Gender', palette = gender_colors)
phd.set_xlim(35,50)
phd.set_ylim(100000,230000)

plt.legend(frameon=False,loc="upper left")
plt.title("Scatterplot: Age vs Salary by Gender among PhD Degree Holders")
plt.show()
```
